# ¿Qué ángulos tomamos como 'pequeños'?

En este notebook analizaremos para qué valores se cumple la aproximación de $sen(\theta)$ $\approx \theta$, respecto a un porcentaje de error de 2%. Se trabajarán con ángulos positivos, pero queda demostrado de igual forma para sus negativos.

```python
#Importamos librerías a utilizar
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generamos los parámetros iniciales, sabiendo que cuando tetta es 0, su seno es 0,
#por lo que esa aproximación es del 100% de certeza.
tetta = 0
tetta_aprox = [] 
```

```python
#Ya que vamos a trabajar con un error del 2%, el valor del seno debe ser mayor 
#o igual al 98% del valor del ángulo 
def f(angle):
    return angle*0.98
    '''
    Esta función calcula el 98% del valor de un ángulo introducido por el usuario
    
    Params:
        angle: Float. El valor del ángulo en radianes.
    
    Returns:
        El 98% del valor de la variable 'angle'.

    '''
```

```python
#Creamos un ciclo while, cuyo objetivo es analizar si el valor del 
#seno de tetta es mayor o igual a su 98%, mientras lo sea, tetta
#se añadirá a la lista de tetta_aprox y se evaluará el ciclo 
#sumándole 0.1 grados sexagesimáles al ángulo anterior, empezando con 0.


while np.sin(tetta) >= f(tetta):
    tetta_aprox.append(tetta)
    tetta = tetta +0.001
    
#Guardamos la lista de valores 'tetta_aprox'
%store tetta_aprox
```

```python
tetta_aprox
```

```python
print('La aproximación con la que estamos trabajando se cumple para ángulos positivos desde '
      ,np.min(tetta_aprox),'hasta',np.max(tetta_aprox))
```

```python
#Podemos intentar graficar estos valores, para demostrar gráficamente
#que en este intervalo de ángulos, la función seno parece lineal
plt.figure(figsize=(10,6))
plt.plot(tetta_aprox,np.sin(tetta_aprox), 'k-')
plt.title(r'Ángulos "pequeños" donde $sen(\theta)$$\approx \theta$', size=17)
plt.xlabel(r'Ángulos ($\theta$) en rad)', size=14)
plt.ylabel(r'Seno de ($\theta$)', size=14)
plt.grid(True)
```

Podemos concluir que para que se cumpla que $sen(\theta)$$\approx \theta$ con un 98% de precisión,
debemos tomar ángulos entre 0 y 0.347, que convertidos a sexagesimales, estarían entre 0 y 20°.
